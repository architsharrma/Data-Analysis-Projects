### Q.10) What are the courses that are published in 2015?

```python
udemy_data.dtypes
```

```python
## NOW AS OUR 'published_timestamp' HAS DATATYPE OBJECT SO FIRSTLY WE HAVE TO CONVERT IT INTO A DATE/TIME FORMAT
udemy_data['published_timestamp']=pd.to_datetime(udemy_data['published_timestamp'])
udemy_data['year'] =udemy_data['published_timestamp'].dt.year
udemy_data[udemy_data.year==2015]
```
